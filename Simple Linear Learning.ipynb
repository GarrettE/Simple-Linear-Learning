Simple Linear Learning

```python
import numpy as np
import cv2
```

Create Labels

```python
labels = ["dog","cat","zebra"]
```

Create Random Seed

```python
np.random.seed(1)
```

W is the wieght matrix, 3 RGB 32 x 32 pixel image
b is the bias vector - to give labels bais

```python
W = np.random.randn(3, 3072)
b = np.random.randn(3)
```

Load image 
flatten image

```python
image = cv2.imread("dog.png")
image_flatten = cv2.resize(image, (32, 32)).flatten()
```

score output = image + bias (3 labels)

```python
outputs = W.dot(image_flatten) + b
```

```python
Output labels
```

```python
for (label,ouput) in zip (labels,outputs):
    print ("info {} : {}".format(label, ouput))
```

choose label with highest output and label image 

```python
cv2.putText(image, "Label: {}".format(labels[np.argmax(outputs)]),
	(10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)   
```

```python
cv2.imshow("Image", image)
cv2.waitKey(0)
```
